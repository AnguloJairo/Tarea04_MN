### Literal a  
### Dibuja las gráficas de y=x−1 y  y=e^(1-x^2)


```python
import numpy as np
import matplotlib.pyplot as plt

# Definir x en el rango [-2, 2]
x = np.linspace(-2, 2, 400)
y1 = x - 1
y2 = np.exp(x)

# Crear la gráfica
plt.plot(x, y1, label='y = x - 1', color='blue')
plt.plot(x, y2, label='y = e^x', color='red')

# Punto aleatorio añadido a x = 0.0001234
x_rand = 0.0001234 + np.random.uniform(-0.1, 0.1)
y_rand1 = x_rand - 1
y_rand2 = np.exp(x_rand)

plt.plot(x_rand, y_rand1, 'bo')
plt.plot(x_rand, y_rand2, 'ro')

# Configurar gráfica
plt.axhline(0, color='gray', linestyle='--')
plt.axvline(0, color='gray', linestyle='--')
plt.title('Gráficas de y = x - 1 y y = e^x')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

### Literal b  
### Usar el método de bisección para encontrar una aproximación con precisión de 10^−5 para un valor de 𝑥 en [−2,0] 

```python
import math

def f(x):
    return x**2 - 1 - math.exp(1 - x**2)

def biseccion(a, b, tol=1e-3):
    # Verificar que hay cambio de signo
    if f(a) * f(b) >= 0:
        print("Error: No hay cambio de signo en el intervalo")
        return None
    
    print(f"\nBuscando raíz en [{a}, {b}] con tolerancia {tol}")
    print("Iter |   a   |   b   |   c   |  f(c)")
    print("-------------------------------------")
    
    iter = 0
    while (b - a) > tol:
        iter += 1
        c = (a + b) / 2
        print(f"{iter:4} | {a:.4f} | {b:.4f} | {c:.4f} | {f(c):.6f}")
        
        if f(c) == 0:
            break
        elif f(a) * f(c) < 0:
            b = c
        else:
            a = c
    
    print("-------------------------------------")
    print(f"Raíz encontrada: x ≈ {c:.5f}")
    return c

# Intervalo dado en el problema
a, b = -2.0, 0.0

# Verificar el intervalo
print(f"f({a}) = {f(a):.4f}")
print(f"f({b}) = {f(b):.4f}")

# Ejecutar bisección
solucion = biseccion(a, b)
```
