### Parte (a): Graficar ùë¶=ùë• y y=sinx

```python
import numpy as np
import matplotlib.pyplot as plt

# Rango de x para graficar
x = np.linspace(0, 2*np.pi, 400)
y1 = x
y2 = np.sin(x)

# Graficar
plt.figure(figsize=(8,5))
plt.plot(x, y1, label=r'$y = x$', linestyle='--')
plt.plot(x, y2, label=r'$y = \sin x$', color='red')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Gr√°ficas de $y = x$ y $y = \sin x$")
plt.grid(True)
plt.legend()
plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linewidth=0.5)
plt.show()
```

### Parte (b): Resolver x=2sinx usando el m√©todo de bisecci√≥n

```python
import math

def f(x):
    return x - 2 * math.sin(x)

def biseccion(f, a, b, tol=1e-5, max_iter=100):
    if f(a) * f(b) >= 0:
        raise ValueError("La funci√≥n no cambia de signo en el intervalo")
    
    for _ in range(max_iter):
        c = (a + b) / 2
        if abs(f(c)) < tol or (b - a) / 2 < tol:
            return c
        if f(a) * f(c) < 0:
            b = c
        else:
            a = c
    return (a + b) / 2

# Intervalo [0, 2] porque gr√°fica muestra que la ra√≠z est√° all√≠
raiz = biseccion(f, 1, 2, tol=1e-5)

print(f"La soluci√≥n aproximada de x = 2 sin(x) es x ‚âà {raiz:.5f}")
```
