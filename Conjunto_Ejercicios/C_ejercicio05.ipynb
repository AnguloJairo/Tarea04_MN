```python
import numpy as np

def f(x):
    return (x + 3) * (x + 1)**2 * x * (x - 1)**3 * (x - 3)

def biseccion_convergencia(a, b, tol=1e-6):
    if np.sign(f(a)) == np.sign(f(b)):
        return None  # Devuelve None cuando no hay convergencia
    
    while (b - a) > tol:
        c = (a + b) / 2
        if f(c) == 0:
            return c
        elif np.sign(f(a)) != np.sign(f(c)):
            b = c
        else:
            a = c
    return (a + b) / 2
```

### Literal a: [-1.5, 2.5]

```python
raiz_a = biseccion_convergencia(-1.5, 2.5)
print(f"a. Intervalo [-1.5, 2.5] converge a: {raiz_a if raiz_a is not None else 'No converge'}")
```

### Literal b: [-0.5, 2.4]

```python
raiz_b = biseccion_convergencia(-0.5, 2.4)
print(f"b. Intervalo [-0.5, 2.4] {'No converge' if raiz_b is None else raiz_b}")
```

### Literal c: [-0.5, 3]

```python
raiz_c = biseccion_convergencia(-0.5, 3)
print(f"c. Intervalo [-0.5, 3] converge a: {raiz_c if raiz_c is not None else 'No converge'}")
```

### Literal d: [-3, -0.5]

```python
raiz_d = biseccion_convergencia(-3, -0.5)
print(f"d. Intervalo [-3, -0.5] converge a: {raiz_d if raiz_d is not None else 'No converge'}")
```
