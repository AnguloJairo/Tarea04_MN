```python
def f(x):
    return x**3 - 7*x**2 + 14*x - 6
```

### Funcion de la Bisección

```python
def biseccion(f, a, b, tol=1e-2, max_iter=100):
    if f(a) * f(b) >= 0:
        raise ValueError("La función no cambia de signo en el intervalo")
    for _ in range(max_iter):
        c = (a + b) / 2
        fc = f(c)
        if abs(fc) < tol or (b - a) / 2 < tol:
            return c
        if f(a) * fc < 0:
            b = c
        else:
            a = c
    return (a + b) / 2
```

```python
# Intervalos dados
intervalos = [(0,1), (1,3.2), (3.2,4)]

for a,b in intervalos:
    raiz = biseccion(f, a, b)
    print(f"Raíz en [{a}, {b}] ≈ {raiz:.4f}")
```
