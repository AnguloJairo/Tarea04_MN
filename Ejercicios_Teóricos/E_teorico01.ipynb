```python
import math

a = 1
b = 2
epsilon = 1e-4

n = math.ceil(math.log2((b - a) / epsilon))
print(f"Número mínimo de iteraciones necesarias: {n}")

def f(x):
    return x**3 - x - 1

def biseccion(a, b, tol=1e-4, max_iter=100):
    if f(a) * f(b) > 0:
        raise ValueError("La función no cambia de signo en el intervalo.")

    for i in range(max_iter):
        c = (a + b) / 2
        fc = f(c)
        print(f"Iteración {i+1}: a = {a:.5f}, b = {b:.5f}, c = {c:.5f}, f(c) = {fc:.5f}")
        
        if abs(fc) < 1e-8 or (b - a) / 2 < tol:
            return c
        
        if f(a) * fc < 0:
            b = c
        else:
            a = c
    
    raise Exception("No se alcanzó la precisión deseada.")

raiz = biseccion(a, b, tol=epsilon)
print(f"\nRaíz aproximada con tolerancia {epsilon}: x = {raiz:.6f}")
```
