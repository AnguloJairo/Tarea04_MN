### Definir la funcin

```python
import math

g = 9.81          
s0 = 300.0        
m = 0.25         
k = 0.1          

def s(t):
    term1 = s0
    term2 = (m * g / k) * t
    term3 = (m**2 * g / k**2) * (1 - math.exp(-k * t / m))
    return term1 - term2 + term3

```

### Mtodo de biseccin

```python
def biseccion(a, b, tol=0.01, max_iter=100):
    if s(a) * s(b) > 0:
        raise ValueError("La funcin no cambia de signo en el intervalo.")
    
    for i in range(max_iter):
        c = (a + b) / 2
        sc = s(c)
        print(f"Iteracin {i+1}: a = {a:.5f}, b = {b:.5f}, c = {c:.5f}, s(c) = {sc:.5f}")
        
        if abs(sc) < 1e-5 or (b - a)/2 < tol:
            return c
        
        if s(a) * sc < 0:
            b = c
        else:
            a = c
    
    raise Exception("No se alcanz la precisin deseada.")
```

```python
a = 0.0
b = 60.0

t_impacto = biseccion(a, b)
print(f"\nTiempo aproximado en que el objeto golpea el suelo: t = {t_impacto:.2f} segundos")
```
