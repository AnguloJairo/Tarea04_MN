```python
import math

# Parmetros del problema
L = 10.0      
r = 1.0       
V = 12.4     

def f(h):
    if h <= 0 or h >= r:
        return 9999  
    term1 = 0.5 * math.pi * r**2
    term2 = r**2 * math.asin(h / r)
    term3 = h * math.sqrt(r**2 - h**2)
    volumen = L * (term1 - term2 - term3)
    return volumen - V

# Mtodo de biseccin
def biseccion(a, b, tol=0.01):
    while (b - a) / 2 > tol:
        c = (a + b) / 2
        if f(c) == 0:
            return c
        elif f(a) * f(c) < 0:
            b = c
        else:
            a = c
    return (a + b) / 2

h = biseccion(0.01, 0.99)
print(f"Profundidad aproximada del agua: h = {h:.4f} cm")
```
