```python
import numpy as np
import matplotlib.pyplot as plt

# Crear valores de x evitando la asíntota de tan(x) en pi/2 (~1.57)
x = np.linspace(0, 1.55, 500)
y1 = x
y2 = np.tan(x)

plt.plot(x, y1, label='y = x', color='blue')
plt.plot(x, y2, label='y = tan(x)', color='red')
#plt.axvline(np.pi/2, color='gray', linestyle='--', label='x = pi/2 (asíntota)')
plt.ylim(0, 5)  # Limitar eje y para que no explote cerca de la asíntota
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfica de y = x y y = tan(x)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import math

def f(x):
    """Función a resolver: f(x) = tan(x) - x"""
    return math.tan(x) - x

def biseccion(a, b, tol=1e-5, max_iter=100, verbose=True):
    # Verificación inicial
    fa, fb = f(a), f(b)
    if fa * fb >= 0:
        raise ValueError(f"No hay cambio de signo en el intervalo [{a}, {b}]. f(a)={fa:.6f}, f(b)={fb:.6f}")
    
    if verbose:
        print("\nMétodo de Bisección")
        print(f"Intervalo inicial: [{a}, {b}]")
        print(f"Tolerancia: {tol}")
        print("-" * 65)
        print(f"{'Iter':<5} | {'a':<10} | {'b':<10} | {'c':<10} | {'f(c)':<12} | {'Error':<10}")
        print("-" * 65)

    for iter in range(1, max_iter + 1):
        c = (a + b) / 2
        fc = f(c)
        error = (b - a) / 2
        
        if verbose:
            print(f"{iter:<5} | {a:<10.6f} | {b:<10.6f} | {c:<10.6f} | {fc:<12.6f} | {error:<10.6f}")

        # Criterios de parada
        if abs(fc) < tol or error < tol:
            if verbose:
                print("-" * 65)
                print(f"Convergencia alcanzada en {iter} iteraciones")
                print(f"Raíz aproximada: {c:.7f} con error < {error:.2e}")
            return c

        # Actualizar intervalo
        if f(a) * fc < 0:
            b = c
        else:
            a = c

    raise Exception(f"No se alcanzó la tolerancia en {max_iter} iteraciones. Mejor aproximación: {c:.7f}")

# Encontrar el primer intervalo positivo válido
def encontrar_intervalo():
    """Busca automáticamente el primer intervalo con cambio de signo"""
    paso = 0.1
    a = 0.0
    while a < 10:  # Límite superior para la búsqueda
        b = a + paso
        try:
            if f(a) * f(b) < 0:
                return a, b
        except ValueError:  # Para evitar singularidades de tan(x)
            pass
        a = b
    raise ValueError("No se encontró un intervalo válido en el rango buscado")

# Ejecución principal
if __name__ == "__main__":
    try:
        # Intervalo para la primera raíz positiva (aproximadamente [4.4, 4.5])
        a, b = 4.4, 4.5
        
        # Verificación del intervalo
        print(f"Verificación del intervalo:")
        print(f"f({a}) = {f(a):.6f}")
        print(f"f({b}) = {f(b):.6f}")
        print(f"Producto f(a)*f(b) = {f(a)*f(b):.6f}\n")
        
        # Ejecutar método de bisección
        raiz = biseccion(a, b)
        

        
    except Exception as e:
        print(f"\nError: {e}")



```
